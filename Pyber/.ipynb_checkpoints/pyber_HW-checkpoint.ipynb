```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_data_df = pd.merge(city_data_df, ride_data_df, on="city")

# Display the data table for preview
combined_data_df.head()
```

```python
city_df = combined_data_df["type"]
city_df.unique()

urban_type = combined_data_df.loc[(combined_data_df["type"] == "Urban")]
suburban_type = combined_data_df.loc[(combined_data_df["type"] == "Suburban")]
rural_type = combined_data_df.loc[(combined_data_df["type"] == "Rural")]


# Obtain the x and y coordinates for each of the three city types

total_ride_urban = urban_type.groupby(["city"]).count()["ride_id"]
urban_avg_fare = urban_type.groupby(["city"])["fare"].mean()
urban_total_drivers = urban_type.groupby(["city"])["driver_count"].value_counts()

total_ride_suburban = suburban_type.groupby(["city"]).count()["ride_id"]
suburban_avg_fare = suburban_type.groupby(["city"])["fare"].mean()
suburban_total_drivers = suburban_type.groupby(["city"])["driver_count"].value_counts()


total_ride_rural = rural_type.groupby(["city"]).count()["ride_id"]
rural_avg_fare = rural_type.groupby(["city"])["fare"].mean()
rural_total_drivers = rural_type.groupby(["city"])["driver_count"].value_counts()

```

```python
# Build the scatter plots for each city types
plt.scatter(total_ride_urban, urban_avg_fare, s=urban_total_drivers*10, 
           marker="o", facecolor="coral", edgecolor="black", alpha=0.5,
           label="Urban")

plt.scatter(total_ride_suburban, suburban_avg_fare, s=suburban_total_drivers*10, 
           marker="o", facecolor="cyan", edgecolor="black", alpha=0.5,
           label="Suburban")

plt.scatter(total_ride_rural, rural_avg_fare, s=rural_total_drivers*10, 
           marker="o", facecolor="gold", edgecolor="black", alpha=0.5,
           label="Rural")

# Incorporate the other graph properties
# Create a legend
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.legend(loc="upper right")

# Incorporate a text label regarding circle size
plt.text(42,35, "Note:\nCircle size correlates with driver count per city")
plt.grid(True)

# Save Figure
plt.savefig("images/Pyber_Scatter_plot.png", bbox_inches="tight")
plt.show()
```

```python
# Calculate Type Percents
total_fare_city = combined_data_df.groupby(["type"])["fare"].sum()

# Build Pie Chart
labels=["Rural","Suburban","Urban" ]
colors=["gold", "cyan", "coral"]
explode = (0,0,0.1)

plt.pie(total_fare_city, colors=colors, labels=labels, explode=explode,
       autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Types")
plt.axis("auto")

# Save Figure
plt.savefig("images/Pyber_Pie_Total_Fare.png", bbox_inches="tight")
plt.show()
```

```python
# Calculate Ride Percents
total_rides_city = combined_data_df.groupby(["type"])["ride_id"].count()

# Build Pie Chart
labels=["Rural","Suburban","Urban" ]
colors=["gold", "cyan", "coral"]
explode = (0,0,0.1)

plt.pie(total_rides_city, colors=colors, labels=labels, explode=explode,
       autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Rides by City Types")
plt.axis("auto")

# Save Figure
plt.savefig("images/Pyber_Pie_Total_Rides.png", bbox_inches="tight")
plt.show()
```

```python
# Calculate Driver Percents
total_drivers_city = city_data_df.groupby(["type"])["driver_count"].sum()

# Build Pie Charts
labels=["Rural","Suburban","Urban" ]
colors=["gold", "cyan", "coral"]
explode = (0,0,0.1)

plt.pie(total_drivers_city, colors=colors, labels=labels, explode=explode,
       autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Drivers by City Types")
plt.axis("auto")

# Save Figure
plt.savefig("images/Pyber_Pie_Total_drivers.png", bbox_inches="tight")
plt.show()
```
